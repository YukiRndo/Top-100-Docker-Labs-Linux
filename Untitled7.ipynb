```python
import numpy as np

print('Uniform')
print(np.random.rand(5))

print('\nNormal')
print(np.random.randn(5))

print('\nArange')
print(np.arange(-10,10,2))

print('\nReshape')
print(np.arange(12).reshape(3,4))

print('\nLinspace')
print(np.linspace(0,1,10))

print('\nLogspace')
print(np.logspace(-3,3,7))

print('\nZeros')
print(np.zeros((2,3)))

print('\nOnes')
print(np.ones((3,2)))

print('\nIdentity')
print(np.eye(3))
```
