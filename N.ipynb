```python
from pandas import DataFrame

cars = {
'make': ['Ford', 'Honda', 'Toyota', 'Tesla'],
'model': ['Taurus', 'Accord', 'Camry', 'Model S'],
'MSRP': [27595, 23570, 23495, 68000]
}

carData = DataFrame(cars)

carData

print('carData.index = ', carData.index)

print('carData.columns = ', carData.columns)

carData2 = DataFrame(cars, index = [1,2,3,4])

carData2['year'] = 2018

carData2['dealership'] = [
'Courtesy Ford',
'Capital Honda',
'Spartan Toyota',
'North Tesla'
]

carData2
```

```python
from pandas import Series
import numpy as np

s2 = Series(np.random.randn(6))

print('Series s2 =\n', s2, '\n')

print('s2.values = ', s2.values)

print('s2.index = ', s2.index)

print('s2.dtype = ', s2.dtype)
```

```python
import numpy as np

print('Uniform')
print(np.random.rand(5))

print('\nNormal')
print(np.random.randn(5))

print('\nArange')
print(np.arange(-10,10,2))

print('\nReshape')
print(np.arange(12).reshape(3,4))

print('\nLinspace')
print(np.linspace(0,1,10))

print('\nLogspace')
print(np.logspace(-3,3,7))

print('\nZeros')
print(np.zeros((2,3)))

print('\nOnes')
print(np.ones((3,2)))

print('\nIdentity')
print(np.eye(3))
```

```python
import numpy as np

oneDim = np.array([1.0,2,3,4,5])

print(oneDim)

print("#Dimensions =", oneDim.ndim)

print("Dimension =", oneDim.shape)

print("Size =", oneDim.size)

print("Array type =", oneDim.dtype)
```
